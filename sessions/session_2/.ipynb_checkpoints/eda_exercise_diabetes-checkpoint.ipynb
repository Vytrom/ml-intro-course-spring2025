### Exercise 4: Handling Missing Data (Implicit Missing Values)

As observed in Exercise 1.3, some columns have a minimum value of 0, which might represent missing data (e.g., a blood pressure of 0 is biologically impossible).

1. For the `Glucose`, `BloodPressure`, `SkinThickness`, `Insulin`, and `BMI` columns, replace the 0 values with `NaN` (Not a Number) to explicitly mark them as missing.
2. Verify the number of missing values in these columns after the replacement using `.info()` or `.isnull().sum()`. Which column has the most missing values now?
3. Choose an imputation strategy for the missing numerical values (e.g., replace with the mean or median of the respective column). Justify your choice.
4. Apply the chosen imputation strategy to fill the `NaN` values in the affected columns. Verify that there are no more missing values in these columns.

```python
# Your code for Exercise 4.1
mod_list =  ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for x in mod_list:
    mod_diabetes_df = diabetes_df
    mod_diabetes_df[x] = diabetes_df[x].replace(0,None)
mod_diabetes_df.info()
```

```python
# Your code for Exercise 4.2
mod_diabetes_df.info()
```

```python
# Your code for Exercise 4.3 and 4.4(Add comments for justification)

'''Glucose, insulin and SkinThickness are diabetes-depedent. BMI and BP are not too correlated, and therefore a higher sample is more reliable. 
SkinThickness has a huge outlier, so median should be more accurate.'''
G0 = diabetes_df[diabetes_df['Outcome']==0].Glucose.mean()
G1 = diabetes_df[diabetes_df['Outcome']==1].Glucose.mean()
ST0 = diabetes_df[diabetes_df['Outcome']==0].SkinThickness.median()
ST1 = diabetes_df[diabetes_df['Outcome']==1].SkinThickness.median()
I0 = diabetes_df[diabetes_df['Outcome']==0].Insulin.mean()
I1 = diabetes_df[diabetes_df['Outcome']==1].Insulin.mean()
BMI = diabetes_df.BMI.mean()
BP = diabetes_df.BloodPressure.mean()

diabetes_df[diabetes_df['Outcome']==0].fillna({"Glucose": G0}, inplace=True)
diabetes_df[diabetes_df['Outcome']==0].fillna({"SkinThickness": ST0}, inplace=True)
diabetes_df[diabetes_df['Outcome']==0].fillna({"Insulin": I0}, inplace=True)

diabetes_df[diabetes_df['Outcome']==1].fillna({"Glucose": G1}, inplace=True)
diabetes_df[diabetes_df['Outcome']==1].fillna({"SkinThickness": ST1}, inplace=True)
diabetes_df[diabetes_df['Outcome']==1].fillna({"Insulin": I1}, inplace=True)

diabetes_df.fillna({"BMI": BMI}, inplace=True)
diabetes_df.fillna({"BloodPressure": BP}, inplace=True)

diabetes_df.info()
```

### Exercise 5: Creating New Features (Feature Engineering)

Create a new categorical feature based on the `Age` column.

1. Define age bins (e.g., '0-20', '21-30', '31-40', '41-50', '51+').
2. Create a new column called `AgeGroup` by categorizing the `Age` values into these bins.
3. Display the value counts for the new `AgeGroup` column.

```python
# Your code for Exercise 5.1, 5.2, and 5.3
# Definir los bins y etiquetas para los grupos de edad
age_bins = [0, 20, 30, 40, 50, float('inf')]  # float('inf') captura todo lo mayor a 50
age_labels = ['0-20', '21-30', '31-40', '41-50', '51+']

# Crear la nueva columna 'AgeGroup' usando pandas.cut()
diabetes_df['AgeGroup'] = pd.cut(
    diabetes_df['Age'],
    bins=age_bins,
    labels=age_labels,
    right=False  # Incluye el límite inferior, pero no el superior (ej: 20 está en '0-20', no en '21-30')
)

# Mostrar el conteo de valores para cada grupo de edad
agegroup_counts = diabetes_df['AgeGroup'].value_counts().sort_index()
print("Conteo de pacientes por grupo de edad:\n", agegroup_counts)
```

### Exercise 6: Exploring Correlations

1. Compute the correlation matrix for the numerical features in the DataFrame using `.corr()`.
2. Display the correlations of all numerical features with the `Outcome` variable, sorted in descending order. Which features have the strongest positive and negative correlations with diabetes outcome?
3. Use seaborn's `heatmap` to visualize the correlation matrix. Interpret the relationships between different features.

```python
# Your code for Exercise 6.1
# Calculate correlation matrix for numerical features
corr_matrix = diabetes_df.select_dtypes(include=['int64', 'float64']).corr()
```

```python
# Your code for Exercise 6.2
outcome_corr = corr_matrix['Outcome'].sort_values(ascending=False)
print("Correlations with Outcome:\n", outcome_corr)

# Get top positive and negative correlations
top_positive = outcome_corr.index[1]  # Skipping Outcome itself
top_negative = outcome_corr.index[-1]

print(f"\nStrongest positive correlation with Outcome: {top_positive} ({outcome_corr[top_positive]:.2f})")
print(f"Strongest negative correlation with Outcome: {top_negative} ({outcome_corr[top_negative]:.2f})")
```

```python
# Your code for Exercise 6.3
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            center=0,
            vmin=-1, vmax=1,
            linewidths=0.5)
plt.title("Correlation Matrix of Diabetes Dataset")
plt.tight_layout()
plt.show()
```

### Exercise 7: Analyzing Relationships with Outcome

Use visualizations to explore the relationship between different features and the `Outcome` (diabetes positive or negative).

1. Create a violin plot using seaborn to visualize the distribution of `Age` for both `Outcome` groups (0 and 1).
2. Create a scatter plot of `Glucose` vs `BMI`, coloring the points based on the `Outcome` variable.
3. Use seaborn's `boxplot` to compare the distribution of `BloodPressure` across different `AgeGroup` categories (created in Exercise 5).

```python
# Your code for Exercise 7.1
diabetes_df['Outcome'] = diabetes_df['Outcome'].astype(int)  # Convert to int

plt.figure(figsize=(10, 6))
sns.violinplot(
    data=diabetes_df,
    x='Outcome',
    y='Age',
    inner='quartile',
    split=True
)
plt.title('Age Distribution by Diabetes Outcome')
plt.xticks([0, 1], ['Non-Diabetic (0)', 'Diabetic (1)'])
plt.xlabel('Outcome')
plt.ylabel('Age (years)')
plt.show()
```

```python
# Your code for Exercise 7.2
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diabetes_df, x='Glucose', y='BMI', 
                hue='Outcome', palette={0: 'blue', 1: 'red'}, 
                alpha=0.7, size='Age', sizes=(20, 200))
plt.title('Glucose vs BMI Colored by Diabetes Outcome')
plt.xlabel('Glucose Level')
plt.ylabel('Body Mass Index (BMI)')
plt.legend(title='Outcome', labels=['Non-Diabetic', 'Diabetic'])
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Your code for Exercise 7.3
plt.figure(figsize=(12, 6))
sns.boxplot(data=diabetes_df, x='AgeGroup', y='BloodPressure', 
            order=['0-20', '21-30', '31-40', '41-50', '51+'],
            palette='Blues', showmeans=True,
            meanprops={'marker':'o', 'markerfacecolor':'white', 'markeredgecolor':'black'})
plt.title('Blood Pressure Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Blood Pressure (mm Hg)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

### Exercise 8: Outlier Identification (Visualization Based)

Use box plots to visually identify potential outliers in some of the key numerical features.

1. Create box plots for `Glucose`, `BloodPressure`, `Insulin`, and `BMI`.
2. Discuss what outliers might represent in this medical dataset and how they could potentially affect machine learning models.

```python
# Your code for Exercise 8.1
# Set up the plot
plt.figure(figsize=(15, 10))

# Create subplots for each feature
features = ['Glucose', 'BloodPressure', 'Insulin', 'BMI']
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=diabetes_df[feature], color='lightblue', width=0.4)
    plt.title(f'Outliers in {feature}')
    plt.ylabel(feature)
    
plt.tight_layout()
plt.show()
```
